```python
import requests
import json
import time
from datetime import datetime, timedelta

API_KEY_NEWS = "824d71387660481daf9cfb5dd7bda00c"
keywords = f"stock market OR tech stocks OR Nasdaq OR S&P 500 OR economy OR Federal Reserve OR NASDAQ OR S&P 500"

# Aktuelles Datum und vor 3 Tagen berechnen
to_date = datetime.now().strftime("%Y-%m-%d")  
from_date = (datetime.now() - timedelta(days=3)).strftime("%Y-%m-%d")  

url = f"https://newsapi.org/v2/everything?q={keywords}&from={from_date}&to={to_date}&domains=reuters.com,bloomberg.com,cnbc.com,wsj.com,ft.com&language=en&apiKey={API_KEY_NEWS}"

response = requests.get(url)
if response.status_code == 200:
    data = response.json()

    file_path = f"/lakehouse/default/Files/news_data.json"

    with open(file_path, "w") as json_file:
        json.dump(data, json_file, indent=4)

else:
    print(f"Fehler: {response.status_code} - {response.text}")

time.sleep(1)
```
